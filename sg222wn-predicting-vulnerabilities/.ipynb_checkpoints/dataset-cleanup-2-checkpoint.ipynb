```python
import pandas as pd
import json
```

```python
with open('soanrqube_vuln.json','r') as f:
    data = json.load(f)
```

```python
data
```

```python
vuln_classes_sonar= []
for i in range(len(data['issues'])):
    if '.java' in data['issues'][i]['component']:
        vuln_classes_sonar.append(data['issues'][i]['component'][30:])
vuln_classes_sonar = set(vuln_classes_sonar)
```

```python
vuln_classes_sonar
```

```python
sonar_df = pd.DataFrame(vuln_classes_sonar, columns=['sonarqube'])
```

```python
len(sonar_df.sonarqube.unique())
```

```python
df= pd.read_csv('cleaned_tomcat_vulnerabilities.csv')
```

```python
df= sonar_df.merge(df, how='inner',left_on='sonarqube'
                   ,right_on='file')
```

```python
df = df[['file','vulnerable','sonarqube']]
```

```python
df =df.fillna(0)
```

```python
vulnerable_df = df.loc[df.vulnerable==1].reset_index(drop=True)
```
