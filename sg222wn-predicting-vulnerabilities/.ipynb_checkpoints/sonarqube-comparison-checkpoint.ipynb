```python
import pandas as pd
```

```python
df = pd.read_csv('cleaned_tomcat_vulnerabilities.csv')
df = df.drop(columns=['Unnamed: 0'])
```

```python
df10=pd.DataFrame(sorted(set(df.loc[(df.version==8)&(df.vulnerable==1)]['class'].values[:])), columns=['apache'])
df10_full = df.loc[df.version==8]
```

```python
len(set(df.loc[(df.version==10)&(df.vulnerable==0)]['class'].values[:]))
```

```python
sonarqube = ['java/org/apache/tomcat/dbcp/jocl/JOCLContentHandler.java',
             'java/org/apache/tomcat/util/DomUtil.java',
            'test/org/apache/tomcat/util/net/TestSsl.java',
             'test/org/apache/tomcat/util/net/TesterSupport.java',
             'java/javax/servlet/http/HttpServlet.java',
             'webapps/examples/WEB-INF/web.xml']
```

```python
sq = pd.DataFrame(sonarqube,columns=['sonarqube'])
```

```python
df10.apache.values[:]
```

```python
df10.merge(sq, left_on='apache',right_on='sonarqube',how='left')
```

```python
inverse = df10_full.merge(sq, left_on='class',right_on='sonarqube',how='left').fillna(0).drop_duplicates(keep='first')
len(set(inverse.loc[(inverse.vulnerable==0)&~(inverse.sonarqube==0)]['class'].values[:]))
```
